```python
import numpy as np
import pandas as pd
import sklearn as sci
from tensorflow import keras as k
from keras.layers import Dense, Activation
from sklearn import preprocessing
```

```python
dataset = pd.read_csv("adult.data", names=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country', 'income'])
dataset
```

```python
encode = preprocessing.LabelEncoder()
# Encode
dataset['sex'] = encode.fit_transform(dataset['sex'].to_numpy()).tolist()
dataset['workclass'] = encode.fit_transform(dataset['workclass'].to_numpy()).tolist()
dataset['education'] = encode.fit_transform(dataset['education'].to_numpy()).tolist()
dataset['marital-status'] = encode.fit_transform(dataset['marital-status'].to_numpy()).tolist()
dataset['occupation'] = encode.fit_transform(dataset['occupation'].to_numpy()).tolist()
dataset['relationship'] = encode.fit_transform(dataset['relationship'].to_numpy()).tolist()
dataset['race'] = encode.fit_transform(dataset['race'].to_numpy()).tolist()
dataset['native-country'] = encode.fit_transform(dataset['native-country'].to_numpy()).tolist()
dataset['income'] = encode.fit_transform(dataset['income'].to_numpy()).tolist()
# Min max scale
dataset['age'] = (dataset['age']-dataset['age'].min())/(dataset['age'].max()-dataset['age'].min()) 
dataset['capital-gain'] = (dataset['capital-gain']-dataset['capital-gain'].min())/(dataset['capital-gain'].max()-dataset['capital-gain'].min()) 
dataset['fnlwgt'] = (dataset['fnlwgt']-dataset['fnlwgt'].min())/(dataset['fnlwgt'].max()-dataset['fnlwgt'].min())
dataset['hours-per-week'] = (dataset['hours-per-week']-dataset['hours-per-week'].min())/(dataset['hours-per-week'].max()-dataset['hours-per-week'].min())
dataset['native-country'] = (dataset['native-country']-dataset['native-country'].min())/(dataset['native-country'].max()-dataset['native-country'].min())
dataset
```

```python
# Перетасовка данных
dataset.reindex(np.random.permutation(dataset.index))
```

```python
x = dataset[
    ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status', 'occupation', 'relationship',
     'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week', 'native-country']].to_numpy()
y = dataset['income'].to_numpy()
```

```python
model = k.Sequential()
model.add(Dense(14, activation = "relu"))
model.add(Dense(7, activation = "sigmoid"))
model.add(Dense(1, activation= "sigmoid"))
model.compile(optimizer='rmsprop',  loss='binary_crossentropy', metrics=['accuracy'])
```

```python
result = model.fit(x, y, epochs=10)
```
