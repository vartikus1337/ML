```python
import numpy as np
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow import keras
from keras.layers import Dense
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
y = df['Species'].to_numpy(copy=True)
```

```python
x = df.drop(columns=['Id', 'Species'], axis=1).to_numpy()
```

```python
x
```

```python
y
```

```python
tokenizer = keras.preprocessing.text.Tokenizer(num_words=4)
tokenizer.fit_on_texts(y)
```

```python
encode_y = tokenizer.texts_to_matrix(y)
encode_y
```

```python
encoded_y = np.delete(encode_y, 0, axis=1)
encoded_y
```

```python
model = keras.Sequential()
```

```python
model.add(Dense(4, activation='sigmoid'))
model.add(Dense(3, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer=keras.optimizers.Adam(0.01),
             metrics=keras.metrics.CategoricalAccuracy()
             )
```

```python
history = model.fit(x, encoded_y, epochs=4)
```

```python
plt.plot(history.history['loss'])
#plt.plot(history.history['categorical_accuracy'])
plt.grid(True) 
plt.show()
```

```python
model.predict([[2.1, 1.5, 2.4, 0.2]])
```
