```python
import requests
import pandas as pd


def get_count_pages_companies():
    json = requests.get('https://habr.com/kek/v2/companies/?page=1&perPage=20&sector=&order=rating&\
                        orderDirection=desc&fl=ru&hl=ru').json()
    return json['pagesCount']


def get_json_page_companies(page_):
    return requests.get(f'https://habr.com/kek/v2/companies/?page={page_}&perPage=20&sector=&order=rating&\
                        orderDirection=desc&fl=ru&hl=ru').json()


def parse_json_company_hubs(alias_):
    return requests.get(f'https://habr.com/kek/v2/companies/{alias_}/investment/hubs').json()


def main():
    max_page = get_count_pages_companies()
    for page in range(1, max_page + 1):
        json_page = get_json_page_companies(page)
        for company_id in json_page['companyIds']:
            alias = json_page['companyRefs'][company_id]['alias']
            name = json_page['companyRefs'][company_id]['titleHtml']  # Имя компании
            subscribers = json_page['companyRefs'][company_id]['statistics']['subscribersCount']  # кол-во подпищеков
            rating = json_page['companyRefs'][company_id]['statistics']['rating']  # кол-во рейтинга
            json_hubs = parse_json_company_hubs(alias)
            for hub_id in json_hubs['hubIds']:
                hub_name = json_hubs['hubRefs'][hub_id]['titleHtml']  # Имя хаба
                hub_impact = json_hubs['hubRefs'][hub_id]['invest']  # Вклад в хаб
                
                
def test1() -> pd.DataFrame:
    data_companies = pd.DataFrame(columns=['name', 'subscribers', 'rating'])
    for page in range(1, 2):
        json_page = get_json_page_companies(page)
        for company_id in json_page['companyIds']:
            name = json_page['companyRefs'][company_id]['titleHtml']
            subscribers = json_page['companyRefs'][company_id]['statistics']['subscribersCount']
            rating = json_page['companyRefs'][company_id]['statistics']['rating']
            data_companies.loc[len(data_companies.index)] = [name, subscribers, rating]
    return data_companies
                

def test2() -> pd.DataFrame:
    json_page = get_json_page_companies(1)   # 1.1 
    company_id = json_page['companyIds'][0]  # 1.2
    name, alias = json_page['companyRefs'][company_id]['titleHtml'], json_page['companyRefs'][company_id]['alias']
    json_hubs = parse_json_company_hubs(alias)
    data_hubs = {}
    for hub_id in json_hubs['hubIds']:
        hub_name, hub_impact = json_hubs['hubRefs'][hub_id]['titleHtml'], json_hubs['hubRefs'][hub_id]['invest']
        data_hubs[hub_name] = [hub_impact]
    data_hubs = pd.DataFrame(data_hubs)
    return data_hubs


if __name__ == '__main__':
    print("def'ы готовы")
```

## P.S 
<h3><code>main()</code> - По моим представления всё должно быть в этой функции </h3>
<h3><code>test1()</code> - Тут парсится name subscribers rating </h3>
<h3><code>test2()</code> - Тут парсится <a href="#hubs">Хабы</a> </h3>

```python
data_companies = test1()
data_hubs = test2()
```
